```python
import re 
from textblob import TextBlob 
import emoji  


import re 
from textblob import TextBlob 
import emoji  
def clean_tweet(tweet): 
    tweet = re.sub(r'@[A-Za-z0-9]+', '', str(tweet)) # remove @mentions
    tweet = re.sub(r'#', '',  str(tweet)) # remove the '#' symbol
    tweet = re.sub(r'RT[\s]+', '',  str(tweet)) # remove RT
    tweet = re.sub(r'https?\/\/S+', '',  str(tweet)) # remove the hyperlink
    tweet = re.sub(r'http\S+', '',  str(tweet)) # remove the hyperlink
    tweet = re.sub(r'www\S+', '',  str(tweet)) # remove the www
    tweet = re.sub(r'twitter+', '',  str(tweet)) # remove the twiiter
    tweet = re.sub(r'pic+', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'com', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'covid-19', '',  str(tweet)) # remove the pic
    #tweet = re.sub(r'covid-19', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'shopping', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'online', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'Shopping', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'Covid-19', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'COVID-19', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'Covid', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'Pandemic', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'pandemic', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'SHOPPING', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'Online', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'ONLINE','',  str(tweet)) # remove the pic
    tweet = re.sub(r'due','',  str(tweet)) # remove the pic
    tweet = re.sub(r'consumer','',  str(tweet)) # remove the pic
    tweet = re.sub(r'amp','',  str(tweet)) # remove the pic
    tweet = re.sub(r'new','',  str(tweet)) # remove the pic
    tweet = re.sub(r'year','',  str(tweet)) # remove the pic
    tweet = re.sub(r'merce','',  str(tweet)) # remove the pic
    tweet = re.sub(r'retail','',  str(tweet)) # remove the pic
    tweet = re.sub(r'covid', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'people', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'store', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'home', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'covid_19', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'shop', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'time', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'delivery', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'sale', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'lockdown', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'covid19', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'coranavirus', '',  str(tweet)) # remove the pic

    return tweet


def remove_emoji(tweet):
    return emoji.get_emoji_regexp().sub(u'', tweet)

```

```python
comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in df2.cleaned_text: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 1000, height = 800, 
                background_color='white', colormap='Set2', 
                collocations=False, 
                stopwords = stopwords, 
                min_font_size = 12).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (10, 10), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
# get functionality of subjectivity and polarity 
def getSubjectivity(text):
    return TextBlob( str(text)).sentiment.subjectivity

def getPolarity(text):
    return TextBlob( str(text)).sentiment.polarity
```

```python
df2.dropna(subset=['cleaned_text'], inplace = True)
df2.reset_index(drop=True, inplace=True)
```

```python
df2['Subjectivity'] = df2['cleaned_text'].apply(getSubjectivity)
df2['Polarity'] = df2['cleaned_text'].apply(getPolarity)
df2.head()
```

```python
def get_Polarity_Analysis(score):
    if score < 0:
      return 'Negative'
    elif score == 0:
      return 'Neutral'
    else:
      return 'Positive'
def get_Subjectivity_Analysis(score):
    if score >  0:
      return 'Opinion'
    else:
      return 'Fact'

df2['Analysis_Polarity'] = df2['Polarity'].apply(get_Polarity_Analysis)

df2['Analysis_Subjectivity'] = df2['Subjectivity'].apply(get_Subjectivity_Analysis)

# Show the dataframe
df2.head(20)

```

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b')
                                #max_df = 0.5, 
                                #min_df = 10)
                               # max_df = 10, 
                               # min_df = .5)
dtm_tf = tf_vectorizer.fit_transform(df2['cleaned_text'].values.astype('U'))
print(dtm_tf.shape)
```

```python
import os
import gensim
from gensim import corpora, models, similarities
import logging
import tempfile
from nltk.corpus import stopwords
from string import punctuation
from collections import OrderedDict
import seaborn as sns
import pyLDAvis.gensim_models as gensimvis
import matplotlib.pyplot as plt
%matplotlib inline

# init_notebook_mode(connected=True) #do not miss this line
```

```python
list1 = ['RT','rt']
stoplist = stopwords.words('english') + list(punctuation) + list1

texts = [[word for word in str(document).lower().split() if word not in stoplist] for document in corpus]

dictionary = corpora.Dictionary(texts)

# print(dictionary)
# print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
tfidf = models.TfidfModel(corpus) # step 1 -- initialize a model
```

```python
corpus_tfidf = tfidf[corpus]  # step 2 -- use the model to transform vectors
print (corpus_tfidf)
```

```python
df_lda = pd.DataFrame(corpus_tfidf)
df_lda = df_lda.fillna(0).T
#print(df_lda.shape)
df_lda.head()
```

```python
total_topics = 5

lda = models.LdaModel(corpus, id2word=dictionary, num_topics=total_topics)
corpus_lda = lda[corpus_tfidf] # create a double wrapper over the original corpus: bow->tfidf->fold-in-lsi

#Show first n important word in the topics:
#lda.show_topics(total_topics,2)
corpus_lda= pd.DataFrame(corpus_lda)
corpus_lda.head(100)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
plt.rcParams["patch.force_edgecolor"] = True
%matplotlib inline
```

```python
#This is another visual representation of Correlation Matrix to use
## SELECT COLORS https://python-graph-gallery.com/92-control-color-in-seaborn-heatmaps/


corrMatrix = df5.corr()

sns.set(font_scale=1)
sns.heatmap(corrMatrix, annot=False,
            xticklabels=corrMatrix.columns,
            yticklabels=corrMatrix.columns,
            cmap='RdBu_r'
            )
plt.title('Correlation matrix')
plt.show()
```

```python
#This is another visual representation of Correlation Matrix to use
sns.set(font_scale=1)
sns.heatmap(corrMatrix, annot=True, linewidth = 0.5, cmap='Spectral_r',square=False, vmin = -1, vmax = 1)
plt.title('Correlation matrix')

plt.show()
```
